```python
from sklearn.preprocessing import MinMaxScaler

# 두 가지 feature를 대상
data = bb#[['length','uppl','godn']]

# 정규화 진행
scaler = MinMaxScaler()
data_scale = scaler.fit_transform(data) + 0.1 #*0.3
data_scale
```

```python
from sklearn.cluster import KMeans

k = 3

# 그룹 수, random_state 설정
model = KMeans(n_clusters = k, random_state = 10)

# 정규화된 데이터에 학습
model.fit(data_scale)

# 클러스터링 결과 각 데이터가 몇 번째 그룹에 속하는지 저장
data['cluster'] = model.fit_predict(data_scale)
data
```

```python
data['cluster'].replace({0 : '군집1',1: '군집2',2:'군집3',3:'군집4',4:'군집5'},inplace =True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (8, 8))

for i in range(3):
    plt.scatter(data.loc[data['cluster'] == '군집'+str(i+1), 'length'], data.loc[data['cluster'] == '군집'+str(i+1), 'difficult'],data.loc[data['cluster'] == '군집'+str(i+1), 'uppl'],data.loc[data['cluster'] == '군집'+str(i+1), 'godn'], 
                label = 'cluster ' + str(i))

plt.legend()
plt.title('K = %d results'%k , size = 15)
plt.xlabel(' ', size = 12)
plt.ylabel(' ', size = 12)
plt.show()
```

```python
from yellowbrick.cluster import KElbowVisualizer

model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,10))  
visualizer.fit(data_scale)
```

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import pandas as pd
km = pd.DataFrame(data = data_scale, 
                      columns = ['length',	'difficult',	'uppl',	'godn'])
pca = PCA(n_components = 2)
pca_transformed = pca.fit_transform(km)
pca_transformed

km['pca_x'] = pca_transformed[:, 0]  #x좌표
km['pca_y'] = pca_transformed[:, 1]  #y좌표
km['cluster'] = data['cluster']
km
```

```python
marker0 = km[km['cluster'] == 0].index
marker1 = km[km['cluster'] == 1].index
marker2 = km[km['cluster'] == 2].index
 
 #scatter plot
plt.scatter(x = km.loc[marker0, 'pca_x'], y = km.loc[marker0, 'pca_y'], marker = 'o')
plt.scatter(x = km.loc[marker1, 'pca_x'], y = km.loc[marker1, 'pca_y'], marker = 's')
plt.scatter(x = km.loc[marker2, 'pca_x'], y = km.loc[marker2, 'pca_y'], marker = '^')
 
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.title('3 Clusters Visualization by 2 PCA Components')
plt.legend(['cluster0', 'cluster1', 'cluster2'])
plt.show()
```
